```python
'''
4- Faça um programa que solicita uma quantidade indefinida de notas de uma prova (deve parar quando um número negativo for digitado).
Armazena em uma lista. Após isso, o programa deve mostrar a média das notas.
'''

notas = []

print("Digite as notas. Digite um número negativo para parar a entrada.")

while True:
    try:
        entrada = float(input("Digite uma nota: "))
        if entrada < 0:
            break
        notas.append(entrada)
    except ValueError:
        print("Entrada inválida. Por favor, digite um número.")
        continue

if len(notas) > 0:
    soma_notas = sum(notas)
    quantidade_notas = len(notas)
    media = soma_notas / quantidade_notas

    print("\n--- Resultados ---")
    print(f"Notas digitadas: {notas}")
    print(f"Total de Notas: {quantidade_notas}")
    print(f"Soma das Notas: {soma_notas:.2f}")
    print(f"Média das Notas: {media:.2f}")
else:
    print("\nNenhuma nota válida foi digitada.")
```

```python
'''
5- Faça um programa que solicita uma quantidade indefinida de números positivos (deve parar quando um número negativo for digitado).
Armazene-os em duas listas: uma para os números pares; outra para os números ímpares.
A seguir, mostre a porcentagem de pares e ímpares digitados.
'''

pares = []
impares = []
print("Digite números positivos. Digite um número negativo para parar.")

while True:
    try:
        entrada = int(input("Digite um número: "))
        if entrada < 0:
            break
        if entrada % 2 == 0:
            pares.append(entrada)
        else:
            impares.append(entrada)

    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")
        continue

total_numeros = len(pares) + len(impares)
if total_numeros > 0:
    porcentagem_pares = (len(pares) / total_numeros) * 100
    porcentagem_impares = (len(impares) / total_numeros) * 100

    print("\n--- Resultados ---")
    print(f"Números Pares: {pares}")
    print(f"Números Ímpares: {impares}")
    print(f"Total de números válidos: {total_numeros}")
    print(f"Porcentagem de Pares: {porcentagem_pares:.2f}%")
    print(f"Porcentagem de Ímpares: {porcentagem_impares:.2f}%")
else:
    print("\nNenhum número positivo foi digitado.")
```

```python
'''
6- Faça um programa que solicita uma quantidade indefinida de números positivos (deve parar quando um número negativo for digitado).
Armazene todos os números digitados em uma lista, sem repetição.
'''

conjunto_numeros = set()
print("Digite números positivos. Digite um número negativo para parar.")

while True:
    try:
        entrada = int(input("Digite um número: "))
        if entrada < 0:
            break
        conjunto_numeros.add(entrada)

    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")
        continue

lista_sem_repeticao = list(conjunto_numeros)

print("\n--- Resultados ---")
print(f"Números Únicos Digitados (em lista): {lista_sem_repeticao}")
```

```python
'''
7- Escreva um programa que solicita que o usuário digite a altura e o peso de 5 pessoas.
Verifique se pelo menos duas das pessoas tem a mesma altura e mesmo peso.
'''

dados_pessoas = []
NUM_PESSOAS = 5

for i in range(NUM_PESSOAS):
    print(f"\n--- Pessoa {i + 1} ---")
    while True:
        try:
            altura = float(input("Digite a altura (m): "))
            peso = float(input("Digite o peso (kg): "))
            dados_pessoas.append((altura, peso))
            break

        except ValueError:
            print("Entrada inválida. Por favor, digite números válidos.")

pares_unicos = set(dados_pessoas)

if len(pares_unicos) < NUM_PESSOAS:
    print("\n✅ VERIFICAÇÃO CONCLUÍDA:")
    print(f"HÁ pelo menos duas pessoas com a mesma altura e mesmo peso.")
    print(f"Pares únicos encontrados: {len(pares_unicos)}")
else:
    print("\n❌ VERIFICAÇÃO CONCLUÍDA:")
    print("NÃO há duas pessoas com a mesma altura e mesmo peso (todos os 5 pares são únicos).")
```

```python
'''
8- Faça um programa que solicita que o usuário digite o nome e a idade de diversas pessoas (o programa deve parar quando um nome vazio for informado).
Retorne o nome da pessoa mais velha.
'''

idade_maxima = -1
nome_pessoa_mais_velha = None
print("Digite o nome e a idade das pessoas. Deixe o nome em branco para parar.")

while True:
    nome = input("\nDigite o nome da pessoa: ").strip()
    if not nome:
        print("Nome vazio detectado. Encerrando a entrada de dados.")
        break

    while True:
        try:
            idade_str = input(f"Digite a idade de {nome}: ")
            idade = int(idade_str)
            if idade < 0:
                print("A idade não pode ser negativa. Tente novamente.")
                continue
            break

        except ValueError:
            print("Entrada de idade inválida. Por favor, digite um número inteiro.")

    if idade > idade_maxima:
        idade_maxima = idade
        nome_pessoa_mais_velha = nome
        print(f"-> {nome} é a pessoa mais velha registrada até agora.")

if nome_pessoa_mais_velha is not None:
    print("\n--- Resultado ---")
    print(f"A pessoa mais velha é: **{nome_pessoa_mais_velha}**")
    print(f"Com a idade de: **{idade_maxima} anos**")
else:
    print("\nNenhum dado válido de pessoa foi inserido.")
```

```python
alunos_db = [
    {"matricula": "2025001", "nome": "Carlos Felix", "curso": "Engenharia de software"},
    {"matricula": "2025002", "nome": "Kaiky Alexandre", "curso": "Ciência da Computação"},
    {"matricula": "2025003", "nome": "Vinicius Gonçalves", "curso": "Sistemas de Informação"},
]

print("1. Sistema Simplificado de Cadastro de Alunos")
print("-" * 40)

# 1. Adicionar Novo Aluno (CREATE)
print("-> Adicionando novo aluno...")
novo_aluno = {"matricula": "2025004", "nome": "Jorge Silva",  "curso": "Redes de Computadores"}
alunos_db.append(novo_aluno)
print(f"Aluno {novo_aluno['nome']} adicionado com sucesso")

#2. Buscar Aluno por Matrícula
matricula_busca = "2025002"
aluno_encontrado = next((aluno for aluno in alunos_db if aluno["matricula"] == matricula_busca), None)

if aluno_encontrado:
  print(f"\nAluno encontrado (Matrícula: {matricula_busca}):")
  print(f"Nome: {aluno_encontrado ['nome']}, Curso: {aluno_encontrado ['curso']}")
else:
  print(f"\nAluno com matrícula {matricula_busca} não encontrado.")

#3. Atualizar Informação do Curso (UPDATE)
matricula_update = "2025002"
for aluno in alunos_db:
  if aluno["matricula"] == matricula_update:
    aluno["curso"] = "Segurança da Informação"
    print(f"\nCurso do aluno {aluno['nome']} atualizado para: {aluno['curso']}")
    break
print("-" * 40)
```

```python
# MINI PROJETO 2: Gerenciamento de Coordenadas (Lista de Tuplas)
# Objetivo: Armazenar sequências de valores fixos (latitude, longitude, nome).
# Lista (locais): Coleção de todas as coordenadas registradas.
# Tupla: Cada tupla (imutável) armazena uma coordenada e seu nome.
locais = [
    # Tupla 1: (latitude, longitude, nome_do_local)
    (-23.55052, -46.63330, "São Paulo"),
    # Tupla 2
    (-15.7801, -47.9292, "Brasília"),
    # Tupla 3
    (40.7128, -74.0060, "Nova York")
]

print("\n2. Gerenciamento de Coordenadas Geográficas")
print("-" * 40)

# 1. Adicionar Nova Coordenada
print("-> Adicionando nova coordenada (Tupla)...")
# Cria uma nova tupla para garantir que a coordenada seja registrada de forma imutável.
novo_local = (34.0522, -118.2437, "Los Angeles")
locais.append(novo_local) # A lista é mutável, permitindo adicionar a nova tupla.
print(f"Local {novo_local[2]} adicionado.")

# 2. Acessar e Desempacotar (Unpack) uma Tupla
print("\n-> Desempacotando a última coordenada...")
# Desempacota os elementos da última tupla da lista em variáveis separadas.
lat, lon, nome = locais[-1]
print(f"Latitude: {lat}, Longitude: {lon}, Local: {nome}")

# 3. Iterar e Formatar Coordenadas
print("\n-> Coordenadas registradas:")
for coordenada in locais:
    # Acessa os elementos da tupla usando índices.
    # coordenada[0] -> latitude, coordenada[1] -> longitude, coordenada[2] -> nome
    print(f"  Local: {coordenada[2]} | Lat/Lon: ({coordenada[0]}, {coordenada[1]})")

# Importante: Tuplas são imutáveis, tentar alterar um elemento (ex: locais[0][0] = 0) resultaria em um erro.
print("-" * 40)
```

```python
# MINI PROJETO 4: Agrupamento de Vendas por Região (Dicionários Aninhados e Tuplas)
# Objetivo: Agrupar dados de vendas (Lista de Tuplas) por uma chave de agrupamento (Região).
# Dicionário (dados_vendas): Chave é a região.
# Valor: Dicionário (dados_venda) contendo detalhes anuais e uma lista de registros de vendas.
dados_vendas = {
    "SUDESTE": {
        "total_anual": 150000.00,
        # Lista de Tuplas: Cada tupla (imutável) é um registro de venda (id_venda, valor)
        "vendas_detalhe": [
            ("V001", 50000.00),
            ("V002", 12000.00),
            ("V003", 8000.00)
        ]
    },
    "SUL": {
        "total_anual": 95000.00,
        "vendas_detalhe": [
            ("V004", 5500.00),
            ("V005", 10000.00)
        ]
    }
}

print("\n4. Agrupamento de Vendas por Região")
print("-" * 40)

# 1. Adicionar Nova Venda a Região SUDESTE
regiao = "SUDESTE"
nova_venda = ("V006", 9500.00)
# Acessa o dicionário SUDESTE, depois a chave 'vendas_detalhe' (que é uma lista), e adiciona a nova tupla.
dados_vendas[regiao]["vendas_detalhe"].append(nova_venda)
print(f"Nova venda {nova_venda[0]} adicionada na região {regiao}.")

# 2. Calcular o Valor Total de Vendas Detalhadas no SUDESTE
vendas_sudeste = dados_vendas["SUDESTE"]["vendas_detalhe"]
# Usa compreensão de lista e a função sum() para somar o segundo elemento (valor) de cada tupla.
soma_detalhes = sum(venda[1] for venda in vendas_sudeste)

print(f"\nSoma dos detalhes de vendas SUDESTE: R$ {soma_detalhes:.2f}")
print(f"Total anual registrado: R$ {dados_vendas[regiao]['total_anual']:.2f}")

# 3. Listar as Regiões (Chaves)
print("\nRegiões em análise:")
# O método .keys() retorna as chaves do dicionário principal.
for reg in dados_vendas.keys():
    print(f"  - {reg}")
print("-" * 40)
```

```python
# MINI PROJETO 5: Mapeamento de Níveis de Acesso (Dicionário com Tuplas)
# Objetivo: Definir conjuntos de permissões fixas por nível de usuário.
# Dicionário (permissoes): Chave é o nível de acesso.
# Tupla: Valor é uma tupla imutável, contendo as permissões (strings).
permissoes = {
    "ADMIN": ("CREATE_USER", "READ_ALL", "UPDATE_ANY", "DELETE_ALL"), # Tupla de permissões para Admin
    "MANAGER": ("READ_ALL", "UPDATE_OWN"), # Tupla de permissões para Manager
    "VIEWER": ("READ_ALL",) # Tupla de um elemento para Viewer (vírgula obrigatória)
}

print("\n5. Mapeamento de Níveis de Acesso e Permissões")
print("-" * 40)

# 1. Acessar Permissões de um Nível Específico (READ)
nivel_acesso = "ADMIN"
# Acessa a tupla de permissões usando a chave do dicionário.
perms_admin = permissoes.get(nivel_acesso)

if perms_admin:
    print(f"Permissões do nível {nivel_acesso}:")
    # Itera sobre a tupla de permissões.
    for perm in perms_admin:
        print(f"  - {perm}")
else:
    print(f"Nível de acesso '{nivel_acesso}' não definido.")

# 2. Verificar uma Permissão Específica
perm_verificar = "DELETE_ALL"
# Usa o operador 'in' para checar se a permissão existe dentro da tupla.
if perm_verificar in permissoes["ADMIN"]:
    print(f"\nADMIN possui a permissão: {perm_verificar}")
else:
    print(f"\nADMIN NÃO possui a permissão: {perm_verificar}")

# 3. Tentativa de Adicionar uma Permissão (Imutabilidade da Tupla)
# print(f"\nTentando adicionar permissão ao VIEWER...")
# permissoes["VIEWER"].append("DELETE_OWN") # Erro: 'tuple' object has no attribute 'append'
# A tupla garante a integridade do conjunto de permissões.
print("\nA integridade do conjunto de permissões é garantida pela Tupla (Imutabilidade).")
```
